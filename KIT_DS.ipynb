```python
def unique_list(l):
  x = []
  for a in l:
    if a not in x:
      x.append(a)
  return x
list1 = [1,2,3,3,3,3,4,5]
print(unique_list(list1))
```

```python
def max_number(a,b,c):
  if a>b and a>c:
    return a
  elif b>a and b>c:
    return b
  else:
    return c
a = int(input("Enter 1st number: "))
b = int(input("Enter 2nd number: "))
c = int(input("Enter 3rd number: "))
print(max_number(a,b,c))
```

```python
def sumlist(l):
  sum = 0
  for i in l:
    sum = sum + i
  return sum
list1 = [1,2,3,3,3,3,4,5]
print(sumlist(list1))
```

```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv('/content/drive/MyDrive/movies-ds.csv')
```

```python
movies = movies.rename(columns = ({'Gross':'Gross in million $'}))
```

```python
movies = movies.drop_duplicates()
```

```python
movies.shape
```

```python
movies = movies.rename(columns = ({'ONE-LINE':'PLOT'}))
```

```python
movies.head(10)
```

```python
movies['YEAR'] = movies['YEAR'].str.strip('-')
```

```python
movies.head(10)
```

```python
movies['YEAR'] = movies['YEAR'].str.strip(' )')
```

```python
movies['YEAR'] = movies['YEAR'].str.strip('(')
```

```python
movies['GENRE'] = movies['GENRE'].str.strip('\n')
```

```python
movies['PLOT'] = movies['PLOT'].str.strip('\n')
```

```python
movies['STARS'] = movies['STARS'].str.strip('\n')
```

```python
movies['STARS'] = movies['STARS'].str.replace('\n','')
```

```python
movies['STARS'] = movies['STARS'].str.replace('     ','')
```

```python
movies['STARS'] = movies['STARS'].str.replace(' Stars', '|')
```

```python
movies[['DIRECTOR' , 'STARS']] = movies['STARS'].str.split('|',1,expand=True)
```

```python
movies['STARS'] = movies['STARS'].str.replace(':','Stars:')
```

```python
movies['STARS'] = movies['STARS'].str.strip('Stars')
```

```python
movies
```

```python
movies.to_excel('Cleaned_Dataset.xlsx',index=False)
```
